## 1.例子-什么是好的算法/程序

例1，写程序实现一个函数PrintN，使得传入一个正整数为N的参数后，能顺序打印从1到N的全部正整数。

```python
# 循环
def PrintN(n):
    
    # 打印从1到n的正整数
    for i in range(1, n+1):
        print(i, end = " ")
    return
PrintN(3000)
```

```python
# 递归
def PrintN(n):
    if n <= 0:
        return

    # 打印从1到n的正整数
    PrintN(n-1)
    print(n, end = " ")
PrintN(3000)
```

例2，一元多项式的标准表达式：
$ f(x) = a_0 + a_1x + ... + a_nx^n $
。现给定一个多项式的阶数n，并将全体系数
$ \{a_i\}_{i = 0}^{n} $
存放在数组
$ a[] $
里。请写程序计算这个多项式在给定点
$ x $
处的值。

```python
# 循环累计求和
def f1(n, a, x):
    sum = 0
    for i in range(n+1):
        
        sum += a[i] * pow(x, i)
    return sum
f1(100, list(range(111)), 1.2)
```

秦九韶:
$ f(x) = a_0 + x(a_1 + x(...(a_{n-1} + x(a_n))...)) $

```python
def f2(n, a, x):
    sum = 0
    for i in range(n, -1, -1):
        # print(i)
        # print(a[i])
        sum = a[i] + sum * x
    return sum
f2(100, list(range(111)), 1.2)
```

timeit模块提供了一种简单的方法来计算一小段 Python 代码的耗时

```python
from timeit import repeat
def f(n):
    for i in range(n):
        print(i)
# repeat(Stmt, number = N，repeat = M, globals = globals())
# 执行Stmt(string)
# number:给定代码片段Stmt执行的次数
# repeat:repeat() 执行的次数,每一次都会有一个输出，记录所耗时间
# globals：指定用于执行代码的命名空间
print(repeat('f(10)', repeat = 2, number = 1, globals = globals()))
```

## 三：应用实例：最大子列和问题

问题：给定
$ n $
个整数的序列
$ \{a_1, a_2, ...,a_n\}$
求函数
$ f(i, j) = max\{\sum_{k=i}^{j}a_k\}$
的最大值。
“子列”被定义为原始序列中连续的一段数字。例子：
$ [-1, 3, -2, 4, -6, 1, 6, -1]$

### a.穷举-- 时间复杂度
$ T(n) = n^3$

```python
# 穷举所有子列和，从中找出最大值
def MaxSubseqSum1(l: list, n: int) -> int:

    maxsum = l[0] # 最大值
    cursum = 0 # 当前和


    for i in range(n): # i是子列左端的位置
        for j in range(i, n): # j是子列右端的位置
            cursum = 0 # cur是从list[i]到list[j]de 子列和
            for k in range(i, j + 1):
                cursum += l[k]
            if (cursum > maxsum): # 如果刚得到的这个子列和更大
                maxsum = cursum   # 更新结果


    return maxsum
```

### b.部分存储中间值的穷举 --时间复杂度
$ T(n) = n^2 $

```python
def MaxSubseqSum2(l: list, n: int) -> int:

    maxsum = l[0] # 最大值
    cursum = 0 # 当前和


    for i in range(n): # i是子列左端的位置
        cursum = 0 # cursum是从list[i]到list[j]de 子列和
        for j in range(i, n): # j是子列右端的位置
            cursum += l[j]
            if (cursum > maxsum): # 如果刚得到的这个子列和更大
                maxsum = cursum   # 更新结果


    return maxsum
```

### c.分治法 -- 时间复杂度
$ T(n) = nlogn $

分而治之：将原问题拆分成若干小型问题，分别解决后再将结果合而治之，递归
 把原始序列一分为二，那么最大子序列或者在左半边，或者在右半边，或者是横跨中分线的一段
）

###### step1：将序列从中分为左右两个子序列
###### step2：递归求得两子列的最大和
###### step3：从中分点分头向左，右两边扫描，找出跨过分界线的最大子列和
###### step4：返回三个和中的最大者

```python
# 递归的写法
# step1：终止条件
# step2：divide
# step3：conquer
def DivideAndConquer(l, left, right):
    if left == right: # 递归的终止条件，子列只有1个数字时
        return l[left]

    # divide
    mid = (left + right) // 2

    MaxLeftSum = DivideAndConquer(l, left, mid)
    MaxRightSum = DivideAndConquer(l, mid+1, right)

    # 求跨分界线的最大子列和
    MaxLeftBorderSum = l[mid]
    LeftBorderSum = 0
    # 从mid开始扫描，知道left结束
    for i in range(mid, left - 1, -1):
        LeftBorderSum += l[i]
        if LeftBorderSum > MaxLeftBorderSum:
            MaxLeftBorderSum = LeftBorderSum

    MaxRightBorderSum = l[mid+1]
    RightBorderSum = 0
    # 从mid开始扫描，知道left结束
    for i in range(mid+1, right+1):
        RightBorderSum += l[i]
        if RightBorderSum > MaxRightBorderSum:
            MaxRightBorderSum = RightBorderSum

    return max(MaxLeftSum, MaxRightSum, MaxRightBorderSum + MaxLeftBorderSum)

def MaxSubseqSum3(l:list, n: int) -> int:
    return DivideAndConquer(l, 0, n-1)
```

### d.在线处理：
如果整数序列
$ \{a_1, a_2, ..., a_n\}$
的最大和子列是
$ \{a_i, a_{i+1}, ..., a_j\}$
，那么必定有
$\sum_{k=i}^{l}a_k\geq0$
对任意
$i\leq l \leq j$
成立。因此，一旦发现当前子列和为负，则可以开始重新考察一个新的子列。

```python
maxsum = l[0] # 记录当前最大值
cursum = 0 # 记录子列和，当cursum小于0的时候，抛弃之前的序列，cursum重新赋值为0
```

```python
# 在线处理
def MaxSubseqSum4(l:list, n:int) -> int:
    maxsum = l[0]
    cursum = 0

    for i in range(1, n):
        cursum += l[i] # 向右累加
        if cursum > maxsum:
            maxsum = cursum

        if cursum < 0:
            cursum = 0
    return maxsum
```
